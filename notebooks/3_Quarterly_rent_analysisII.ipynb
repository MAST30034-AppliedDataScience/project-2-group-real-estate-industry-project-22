### Finding the log average of median monthly rent for each suburb from 2000 - 2024

```python
import numpy as np 
def log_average(df):
    

    # Initialize an empty list to store log_mean values
    log_mean_list = []

    # Convert DataFrame to a list of lists
    rows_list = df.iloc[:, 1:].values.tolist()

    for rows in rows_list:
        series = pd.to_numeric(rows, errors='coerce')
        series = series.astype(float)

        log_values = np.log(series)
        log_mean = np.mean(log_values)
        log_mean_list.append(log_mean)

    # Create a new DataFrame with the log_mean data
    log_mean_df = pd.DataFrame(log_mean_list)
    return log_mean_df


m_sub_f1b = log_average(m_flat_1b)
m_sub_f2b = log_average(m_flat_2b)
m_sub_f3b = log_average(m_flat_3b)

m_sub_h2b = log_average(m_house_2b)
m_sub_h3b = log_average(m_house_3b)
m_sub_h4b = log_average(m_house_4b)

m_sub_all = log_average(m_all_props)
m_sub_df = pd.DataFrame(m_all_props['suburbs'])

m_sub_df.reset_index(drop=True, inplace=True)
m_sub_f1b.reset_index(drop=True, inplace=True)
m_sub_f2b.reset_index(drop=True, inplace=True)
m_sub_f3b.reset_index(drop=True, inplace=True)
m_sub_h2b.reset_index(drop=True, inplace=True)
m_sub_h3b.reset_index(drop=True, inplace=True)
m_sub_h4b.reset_index(drop=True, inplace=True)
m_sub_all.reset_index(drop=True, inplace=True)


m_sub_df = pd.concat([m_sub_df,m_sub_f1b,m_sub_f2b,m_sub_f3b, 
                        m_sub_h2b,m_sub_h3b,m_sub_h4b,
                        m_sub_all],
                        keys=['suburbs','log_mean_1B_flats','log_mean_2B_flats','log_mean_3B_flats',
                        'log_mean_2B_houses', 'log_mean_3B_houses','log_mean_4B_houses',
                        'log_mean_all_properties'], axis =1)
m_sub_df.head()


                        
# test = m_sub_df[m_sub_df['suburbs'] == 'Albert Park-Middle Park-West St Kilda']
# print(test)
```

```python
def split_and_expand(df, column_to_split, delimiter='-'):
    # Split the suburb column 
    df_split = df[column_to_split].str.split(delimiter, expand=True)
    
    # Melt the split columns into a long format
    df_melted = pd.DataFrame(df_split.stack(), columns=[column_to_split]).reset_index(level=1, drop=True)
    
   
    df_expanded = df_melted.merge(df.drop(columns=[column_to_split]), left_index=True, right_index=True)
    
    return df_expanded


df_expanded = split_and_expand(df, 'suburbs')

print(df_expanded)
```
