```python
import pandas
import newscrape
```

```python
postcodes = pandas.read_csv('../data/raw/aus_postcodes.csv')
```

```python
features = ['postcode', 'locality']
postcodes = postcodes[postcodes['state']=='VIC']
vic_codes = postcodes[features]
vic_codes = vic_codes.drop_duplicates('postcode')
vic_codes = vic_codes.drop_duplicates('locality')
# saving the dataframe
vic_codes.to_csv('../data/raw/postcodes.csv')
```

```python
import pandas as pd

# Step 1: Read the JSON file into a Pandas DataFrame
df = pd.read_json('../data/raw/example.json')

# Step 2: Convert any list columns to strings
df = df.applymap(lambda x: ', '.join(x) if isinstance(x, list) else x)

# Step 3: Convert DataFrame to Parquet file
df.to_parquet('example.parquet', engine='pyarrow')

# Step 4: Display the first 10 rows
print("First 10 rows of the DataFrame:")
print(df.head(10))
```
