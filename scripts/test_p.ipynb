```python
from pyspark.sql import SparkSession

# Create a spark session (which will run spark jobs)
spark = (
    SparkSession.builder.appName("MAST30034 Tutorial 1")
    .config("spark.sql.repl.eagerEval.enabled", True) 
    .config("spark.sql.parquet.cacheMetadata", "true")
    .config("spark.sql.session.timeZone", "Etc/UTC")
    .getOrCreate()
)

# immutable : once u create a df w data structure, the contents in the df cannot be changed
# any modifications on the df will create a new df
# csv stored in rows, parquet is stored in columns. 
# harder to index csv cause 1 row has lots of datatypes, 
# parquet : each column has the same datatype & cols are stored tgthr
```

```python
import os

    # from the current `tute_1` directory, go back two levels to the `MAST30034` directory
output_relative_dir = '../data/'

# check if it exists as it makedir will raise an error if it does exist
if not os.path.exists(output_relative_dir):
    os.makedirs(output_relative_dir)
    
# now, for each type of data set we will need, we will create the paths
if not os.path.exists(output_relative_dir + 'scraped'):
    os.makedirs(output_relative_dir + 'scraped')
```
